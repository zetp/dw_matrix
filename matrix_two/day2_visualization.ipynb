```
def group_n_barplot(df, feat_groupby,
                    feat_agg='price_value',
                    agg_funs=[np.mean, np.median, np.size],
                    feat_sort="mean",
                    top=60,
                    subplots=True):
  return(
        df
       .groupby(feat_groupby)[feat_agg]
       .agg(agg_funs)
       .sort_values(by=feat_sort, ascending=False)
       .head(top)
  ).plot(kind='bar', figsize=(15,8), subplots=subplots)
```

```
group_n_barplot(df, 'param_marka-pojazdu');
```

```
group_n_barplot(df, 'param_year', feat_sort="size");
```

```
# 'param_country-of-origin'
# 'param_color'
group_n_barplot(df, 'param_kraj-pochodzenia');
```
